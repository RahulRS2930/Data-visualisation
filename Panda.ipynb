```python
from io import StringIO,BytesIO
```

```python
data=('col1,col2,col3\n'
     'x,y,1\n'
     'a,b,2\n'
     'c,d,3')
```

```python
type(data)
```

```python
df=pd.read_csv(StringIO(data))
df.head()
```

```python
df.to_csv('Test2.csv')
```
