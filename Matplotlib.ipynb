# MatplotLib
## It is a plotting library for python and its numerical mathematics extension numpy.It provides an object oriented API for embedding plots

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.scatter(x,y,c='g') #c defines color of plot
plt.xlabel('X axis') #gives name to x axis
plt.ylabel('Y axis') #name to y axis
plt.title('Graph in 2D') #title of graph
plt.savefig('Test.png') #saves the plot in Test.png file
```

```python
plt.scatter(x,y,c='r')
#as we have already written %matplotlib inline so we need not to write plt.show() to view the plot
```

```python
y=x*x
plt.scatter(x,y,c='b')
```

```python
# plt plot
plt.plot(x,y,'r*-')# we can use different values like linestyle='dashed',linewidth=2,markersize=12
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('using plot')
```

## Creating subplots

```python
plt.subplot(2,2,1) #index 2,2,1 represents 2 rows,2 columns plot and I want this graph at index 1,i.e row 1 column 1
plt.plot(x,y,'r--')  #plots the above subplot at index 1 as specified in subplot
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'b*--')
plt.subplot(2,2,4)
plt.plot(x,y,'bo--')
```

```python
x=np.arange(1,11)
y=3*x + 5
plt.title('Matplotlib demo')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show # not mandatory as we have already used &matplotlib inline
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Matplotlib demo')
plt.plot(x,y,'r')
```

```python
x=np.arange(0,4*np.pi,0.1)
y1=np.sin(x)
y2=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y1,'g')
plt.title('sine')
plt.subplot(2,1,2)
plt.plot(x,y2,'b')
plt.title('cosine')
```

## Bar plot

```python
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

## Histograms

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=20)
plt.title('Histogram')
plt.show()
```

## Box plot

```python
data=[np.random.normal(0,std,100) for std in range(1,4)] #std represents standard deviation value
#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)] #std represents standard deviation value
#rectangular box plot
plt.boxplot(data,vert=False,patch_artist=False); #making artist false removes colour
#vert=false makes plot horizontal
```

## Pie chart

```python
labels='Python','C++','Java','Ruby'
sizes=[215,130,210,245]
colors=['green','yellow','red','blue']
explode=(0.4,0,0,0) #explode 1st slice and 0.4 represents how far the exploded slice will go
#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('Equal')
```
